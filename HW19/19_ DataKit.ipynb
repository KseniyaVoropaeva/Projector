a. Choose a dataset, you can use Seaborn [example datasets]. Create a cheat sheet for yourself containing all plot types discussed in the lecture. Provide the following info:

        - Plot type

```python
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
flights = sns.load_dataset("flights")
```

bar plot

```python
tips = sns.load_dataset("tips")
sns.barplot(x="day", y="total_bill", data=tips)
plt.show()
```

count plot

```python
tips = sns.load_dataset("tips")
sns.countplot(x="day", data=tips)
plt.show()
```

hist plot

```python
tips = sns.load_dataset("tips")
sns.histplot(data=tips, x="total_bill", kde=True)
plt.show()
```

box plot

```python
tips = sns.load_dataset("tips")
sns.boxplot(x="day", y="total_bill", data=tips)
plt.show()
```

scatter plot

```python
tips = sns.load_dataset("tips")
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.show()
```

violin plot

```python
tips = sns.load_dataset("tips")
sns.violinplot(x="day", y="total_bill", data=tips)
plt.show()
```

lm plot

```python
tips = sns.load_dataset("tips")
sns.lmplot(x="total_bill", y="tip", data=tips)
plt.show()
```

pair plot

```python
iris = sns.load_dataset("iris")
sns.pairplot(iris)
plt.show()
```

joint plot

```python
iris = sns.load_dataset("iris")
sns.jointplot(x="sepal_length", y="petal_length", data=iris, kind="scatter")
plt.show()
```

heat map

```python
flights = sns.load_dataset("flights")
flights_pivot = flights.pivot_table(index="month", columns="year", values="passengers")
plt.figure(figsize=(10, 6))  
sns.heatmap(flights_pivot, cmap="YlGnBu", annot=True, annot_kws={"size": 8})  
plt.show()
```
